# importing all shareable only files obtained from previous step

```python
import pandas as pd
import numpy as np
data=pd.read_csv('shareables csv file.csv',encoding='utf8')
data=data.set_index('username')
data
```

# importing mean of likes and comments data

```python
import pandas as pd
datum=pd.read_csv('mean no of likes&comments.csv',encoding='utf8')
datum=datum.set_index('username')
datum
```

# to augment our dataset, above 2 files are joined and result is stored in a new csv file namely 'shareables and mean_data'

```python
data=data.join(datum)
data
```

```python
data.to_csv(r'shareables and mean_data.csv')
```

# The file stored above is imported and another file containing number of followers, following and posts is joined into above file to reach final_dataset 

```python
import pandas as pd
first=pd.read_csv('shareables and mean_data.csv',encoding='utf8')
first=first.set_index('username')
first
```

```python
second=pd.read_csv('no of followers.csv',encoding='utf8')
second=second.set_index('username')
second=second[['FOLLOWERS','FOLLOWING','POSTS']]
second
```

```python
first=first.join(second)
```

# Finally after a join operation the final_dataset is stored to further train our model 

```python
first.to_csv(r'final_dataset.csv')
```
