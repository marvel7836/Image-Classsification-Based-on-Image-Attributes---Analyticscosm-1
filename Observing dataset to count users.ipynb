```python
import csv

# Open details file and get a unique set of links
mean_csv = csv.DictReader(open('mean no of likes&comments.csv','r'))
details = set(i.get('username') for i in mean_csv)

# Open master file and only retain the data not in the set
followers_csv = csv.DictReader(open('no of followers.csv','r'))
master = [i for i in followers_csv if i.get('username') not in details]

# Overwrite master file with the new results
with open('no of followers.csv','w') as file:
    writer = csv.DictWriter(file, master[0].keys())
    writer.writeheader()
    writer.writerows(master)
```
